```python
def getTRA(fileName):

    with open(fileName,'r+') as fp:
        lines = fp.readlines()

    count = 0
    abs_list = []
    trn_list = []
    ref_list = []
    WL_list = []
    for line in lines:
        count += 1
        if count >= 14:
            values = line.split("	")
            WL_list.append(float(values[0])*1000)
            trn_list.append(float(values[1]))
            ref_list.append(float(values[2]))
            abs_list.append(1-float(values[1])-float(values[2]))
            #print(count, line)
    return trn_list,abs_list,ref_list,WL_list
```

```python
trn_list,abs_list,ref_list,WL_list = getTRA("aIGDB_NbTiO2_n0o0V_v3.txt")
ax = plt.subplot()
plt.plot(WL_list,trn_list)
plt.plot(WL_list,abs_list)
plt.plot(WL_list,ref_list)
ax.legend(['T','A','R'])
ax.set_xlabel('wavelength')
ax.set_ylabel('T,R,A')
```

```python
def getTRA_2(fileName):

    with open(fileName,'r+') as fp:
        lines = fp.readlines()
        
    count = 0
    abs_list = []
    trn_list = []
    ref_list = []
    WL_list = []
    for line in lines:
        count += 1
        if count >= 23:
            #print(line)
            values = line.split(",")
            WL_list.append(float(values[0])*1000)
            trn_list.append(float(values[1]))
            ref_list.append(float(values[2]))
            abs_list.append(1-float(values[1])-float(values[2]))
            #print(count, line)
    return trn_list,abs_list,ref_list,WL_list
```
